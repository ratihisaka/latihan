```python
data = [1,2,3,4,5,4,3,2,1,9,4,7]
terbesar = data[0]
for i in range(len(data)):
    if data[i] > terbesar:
        terbesar = data[i]
print(terbesar)
```

```python
data = [1,2,3,4,5,4,3,2,1,-1,-5,9]
terkecil = data[0]
for angka in data:
    if angka < terkecil:
        terkecil = angka
print(terkecil)
```

```python
angka = [1,2,3,4,5,6,3,4,5,7,8,7]
angka_unik = []
for anggota in angka:
    if anggota not in angka_unik:
        angka_unik.append(anggota)
print(angka_unik)
```
