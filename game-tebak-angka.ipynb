```python
import random

def tebak_angka():
    angka_rahasia = random.randint(1, 100)
    tebakan_benar = False
    percobaan = 0

    print("Halo! Saya telah memilih sebuah angka antara 1 dan 100. Coba tebak angka itu!")

    while not tebakan_benar:
        tebakan = int(input("Masukkan tebakan Anda: "))
        percobaan += 1

        if tebakan < angka_rahasia:
            print("Tebakan Anda terlalu rendah. Coba lagi.")
        elif tebakan > angka_rahasia:
            print("Tebakan Anda terlalu tinggi. Coba lagi.")
        else:
            tebakan_benar = True
            print(f"Selamat! Anda menebak angka {angka_rahasia} dengan benar!")
            print(f"Anda berhasil menebak dalam {percobaan} percobaan.")

    jawab = input("Apakah Anda ingin bermain lagi? (ya/tidak): ")
    if jawab.lower() == "ya":
        tebak_angka()
    else:
        print("Terima kasih telah bermain!")

tebak_angka()
```
